## Ambient vibration H/V spectral ratio analysis

### Sensor installation and data acquisition

#### List of required items
- Nanometrics portable broadband seismometers, 
- Trowels and picks
- Compasses
- Handheld GPS

#### Steps-by-step instruction
1. Open the equipment case and check that the Trillium Compact seismometer is not connected.
2. Once at site, set the equipment case near to the test location and remove the GPS antenna. Place in a location with clear view of the sky. It is recommended to do this first because it takes a short periof of time for the GPS receiver to lock onto the satellites. 
3. Dig a small hole of about 10~12cm in diameter, 7~8cm in depth. In grassed areas carefully remove the turf ‘cap’ so it can be put back in place following the test. The hole should be dug to create a fairly flat base.
4. Add a small amount of soil back into the hole to create a workable platform and place the seismometer in the hole and using the compass align the seismometer in a north-south orientation using the arrow on the top of the seismometer 
5. Push the seismometer into the soil softly and check the hole will allow the seismometer to be level once filled it.  Ensure the soil under the sensor is well packed down once level.
6. Fill the gap around the sides of the sensor with soil and tightly pack using hands or a trowel in multiple lifts. Make sure that the seismometer is still level and that the orientation is still north-south.
7. Ensure that there is no long grass, or similar, touching the top of the sensor (that can blow in the wind on the sensor), and if in a site exposed to excessive wind move the equipment case to a location that will help to block this wind.
8. When the installation of seismometer is completed, remove the dust cap and connect the seismometer to the Centaur data logger.
9. Wait until all flashing lights are green and close the equipment case; all the cables need to be routed via designated grooves in the equipment case to avoid damage.
10. Write down the start time and stay away from the equipments. 
11. When the testing is done, write down the end time and disconnect the seismometer from the Centaur data logger trying not to move the seismometer itself until disconnected. Put the dust cap back on the top of the seismometer. 
12. Extract the sensor from the ground, and place all the items back to where they belong. 
13. Do a reasonable job of covering up the effect of the hole that was dug and place turf back on top (use judgement dependent on the quality of the site in the first instance).



### Post-processing the data

#### List of required items
- SD cards from the Nanometrics portable broadband seismometers
- A computer

#### List of required softwares
- Python 2.7 with the Scipy stack and the Obspy package.
- ‘reducemseed.py’ and ‘miniseedReadWriteSlice.py’

The Centaur data loggers at the University of Canterbury are configured to continuously archive the data in the miniseed format and create the data files at the beginning of every hour. When the tests are done at multiple sites within a day, we may keep the Centaur data logger turned on during the day. At the end of the day, we will have multiple data files. These data files need to be merged first, and then split into multiple files which contain only the actual test data.

#### Step-by-step instruction
1. Remove the sd cards from the Centaur data logger.
2. Copy the test data from the SD cards to the local hard drive, keeping them in one folder for each instrument.
3. Copy ‘reducemseed.py’ and ‘miniseedReadWriteSlice.py’ into the data folders. 
4. Create ‘starttime.txt’, and ‘endtime.txt’, which has the start and end time (UTC) in plain text with one line per test. The format of the time is as shown below:
```
yyyymmddhhmmss
```
5. In the command prompt, go to the data folder and run the following command:
```
run reducemseed.py
```

6. This will create three files (one per channel) per test. For example:
```
UC.S01.TEST0.HHE.miniseed
UC.S01.TEST0.HHN.miniseed
UC.S01.TEST0.HHZ.miniseed
```
7. These files can then be used in Geopsy for computing H/V spectral ratio.

### Computing H/V spectral ratio using Geopsy

####  List of required softwares
- Geopsy

#### Step-by-step instruction
1. Load the measured ambient vibration data into Geopsy.
2. Select the XYZ components of measured data and drag to the toolbar button labeled “H/V”.
3. Select the analysis parametres appropriately in the H/V toolbox settings, and start the analysis.

#### Reference: 
http://www.geopsy.org/documentation/geopsy/hv.html
