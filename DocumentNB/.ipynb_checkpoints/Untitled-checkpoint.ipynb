## Ambient vibration H/V spectral ratio analysis

### Sensor installation and data acquisition

#### List of required items
- Nanometrics portable broadband seismometers, 
- Trowels and picks
- Compasses

#### Steps-by-step instruction
1. Open the equipment box and check that the Trillium Compact seismometer should not be connected at this stage.
2. Power on the Centaur data logger if it is not turned on already.
3. Deploy the GPS antenna. It is recommended to do this first because it takes a while until the a GPS receiver locks on the satellites. 
4. Dig a small hole of about 10~12cm in diameter, 7~8cm in depth. Place the Trillium compact seismometer in the hole and check the hole will allow the seismometer to be level once filled it.  Ensure the soil under the sensor is well packed down. 
5. Fill the gap around the sides of the sensor with soils and tightly pack. Precisely align the Trillium Compact to north-south. Make sure that the seismometer is still level.
6. Ensure that there is no long grass, or similar, touching the top of the sensor (that can blow in the wind on the sensor)
7. When the installation of seismometer is completed, connect the seismometer to the Centaur data logger.
8. Make sure all flashing lights are green and close the equipment box; __all the cables need to be routed via designated grooves to avoid damage__.
9. Write down the start time and stay away from the equipments. 
10. When the testing is done, write down the end time and disconnect the seismometer from the Centaur data logger trying not to move the sensor itself untill disconnected.
11. Extract the sensor from the ground, and place all the items back to where they belong. 
12. Do a reasonable job of covering up the effect of the hole that was dug (use judgement dependent on the quality of the site in the first instance)

### Post-processing the data

#### List of required items
- SD cards from the Nanometrics portable broadband seismometers
- A computer

#### List of required softwares
- Python 2.7 with the Scipy stack and the Obspy package.
- ‘reducemseed.py’ and ‘miniseedReadWriteSlice.py’

The Centaur data loggers at the University of Canterbury are configured to continuously archive the data in the miniseed format and create the data files at the beginning of every hour. When the tests are done at multiple sites within a day, we may keep the Centaur data logger turned on during the day. At the end of the day, we will have multiple data files. These data files need to be merged first, and then sliced into multiple files which contain only the actual test data.

#### Step-by-step instruction
1. Remove the sd cards from the Centaur data logger.
2. Copy the test data from the SD cards to the local hard drive, keeping them in one folder for each instrument.
3. Copy ‘reducemseed.py’ and ‘miniseedReadWriteSlice.py’ into the data folders.

### Computing H/V spectral ratio using Geopsy

####  List of required softwares
- Geopsy

#### Step-by-step instruction
1. Load the measured ambient vibration data into Geopsy.
2. Select the XYZ components of measured data and drag to the toolbar button labeled “H/V”.
3. Select the analysis parametres appropriately in the H/V toolbox settings, and start the analysis.

#### Reference: 
http://www.geopsy.org/documentation/geopsy/hv.html
